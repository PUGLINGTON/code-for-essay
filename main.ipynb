```python
import pandas as pd

df = pd.read_csv("slang_dataset.csv")

def classify(comment): #Function to classify how "bro" is used
    text = comment.lower()
    if "thanks bro" in text or "agree bro" in text:
        return "Agreement/Friendliness"
    elif "why" in text or "makes no sense" in text or "stop lying" in text:
        return "Criticism"
    elif "viral" in text or "talented" in text or "smooth" in text:
        return "Praise"
    else:
        return "Humour/Emphasis"

df["Category"] = df["Comment"].apply(classify)#classifying 

print(df)
print("\nCategory Counts:")
print(df.groupby(["Platform", "Category"]).size()) # results
```
